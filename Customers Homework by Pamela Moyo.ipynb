```python
x = ['Prefix','FirstName','LastName','Unnamed: 5','Unnamed: 6','Unnamed: 17','Unnamed: 18','BirthDate.1','Unnamed: 8']
#put the columns with the highest number of empty cells in list x
```

```python
customers.fillna(method = 'ffill', inplace = True)
#using the forward fill method where you use what is at the bottom to fill an empty cell 
#until you encounter a cell with something else and you user the new data to fill empty cells 
#this continues until all cells are filled
```

```python
customers.isna().sum()#checking if there are any empty cells
```

```python
customers.AnnualIncome = customers.AnnualIncome.str.strip('$')
customers.AnnualIncome = customers.AnnualIncome.str.replace(',','')
customers.AnnualIncome#removing signs in the annualincome that make it an object 
```
